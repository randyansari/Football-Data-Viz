```python
import understatapi
from understatapi import UnderstatClient
import pandas as pd
```

```python
understat = understatapi.UnderstatClient()
```

```python
player_data = understat.player(player='227').get_shot_data()
player_data[0]
```

```python
df = pd.DataFrame(player_data)
df.head()
```

```python
filtered_df = df[df['season'] == '2020']
```
